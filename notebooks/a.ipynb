```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load hyperparameter tuning results
base_dir = os.path.abspath(os.path.join(os.getcwd(), ".."))  # Adjust if needed
csv_file_path = os.path.join(base_dir, "results/hyperparameter/hyperparameter_results.csv")
df = pd.read_csv(csv_file_path)

# Display basic statistics and structure of the data
print("Dataset Overview:")
print(df.head())
print(df.describe())

# Visualizing hyperparameter influence on llpd
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="k", y="llpd", hue="lambda", palette="viridis")
plt.xlabel("Hyperparameter k")
plt.ylabel("Log-Likelihood per Data Point (llpd)")
plt.title("Effect of Hyperparameters on llpd")
plt.legend(title="Lambda")
plt.show()
```
